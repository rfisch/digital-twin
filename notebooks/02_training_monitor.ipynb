# 02 — Training Monitor

Monitor MLX LoRA fine-tuning progress: loss curves, adapter info, and sample generation.

```python
import re
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
# Parse training logs (copy output from mlx_lm.lora here or read from file)
# Expected format: "Iter X: Train loss X.XXXX, Val loss X.XXXX"

log_text = """
# Paste your training output here, or read from a log file
"""

train_losses = []
val_losses = []
val_iters = []

for line in log_text.strip().split('\n'):
    train_match = re.search(r'Iter (\d+):.*Train loss (\d+\.\d+)', line)
    val_match = re.search(r'Iter (\d+):.*Val loss (\d+\.\d+)', line)
    
    if train_match:
        train_losses.append((int(train_match.group(1)), float(train_match.group(2))))
    if val_match:
        val_losses.append((int(val_match.group(1)), float(val_match.group(2))))

if train_losses:
    fig, ax = plt.subplots(figsize=(10, 5))
    iters, losses = zip(*train_losses)
    ax.plot(iters, losses, label='Train loss', alpha=0.7)
    if val_losses:
        v_iters, v_losses = zip(*val_losses)
        ax.plot(v_iters, v_losses, label='Val loss', marker='o')
    ax.set_xlabel('Iteration')
    ax.set_ylabel('Loss')
    ax.set_title('Training Progress')
    ax.legend()
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
else:
    print('No training logs to parse. Run training first and paste output above.')
```

```python
# Check adapter checkpoints
adapter_dir = Path('../adapters/jacq-v1')
if adapter_dir.exists():
    files = sorted(adapter_dir.glob('*'))
    print(f'Adapter files in {adapter_dir}:')
    for f in files:
        size_mb = f.stat().st_size / (1024 * 1024)
        print(f'  {f.name}: {size_mb:.1f} MB')
else:
    print(f'{adapter_dir} not found. Run training first.')
```
